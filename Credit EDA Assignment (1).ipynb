```python
for col in cols:
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr = (q3-q1)
    upper_range = q3+(1.5*iqr)
    lower_range = q1-(1.5*iqr)
    df=df.loc[(df[col] > lower_range) & (df[col] < upper_range)]
```

```python
# selecting columns from the application data frame to merge
merge_cols = ['SK_ID_CURR','TARGET','CODE_GENDER','NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS']
df_sub = df[merge_cols]
df_sub.head()
```

```python
# merging previous application data with the subset of the application data frame
merged = pd.merge(df1,df_sub, on='SK_ID_CURR', how='left')
merged.head()
```
